### Question 1
Imagine we have a data frame called `data`, with a `type` column. Which one works and why?\
Function 1:
```
group_and_tally <- function(df, column){
    df %>% group_by({{ column }}) %>% tally();
}
group_and_tally(data, type);
```

Function 2:
```
group_and_tally <- function(df, column){
    df %>% group_by(column) %>% tally();
}
group_and_tally(data, type);
```

Answer: 
Function 1 works in this case because the double curly brackets indicate to call by the name of the column versus by the word "column" as the name of a column. In this case, the column name is "type" and column is just the type of organizer. Function 2 would not work because it indcates to organize and tally by the word column versus type.
