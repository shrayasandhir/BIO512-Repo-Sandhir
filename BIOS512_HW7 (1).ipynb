## Question 2
#### Explain how patches relate to the idea of committing a change to the history of your code?

Patches describes any changes that have been made to code by recroding differences between one version and another. This can be a modification, a removal, or an addition. While the patch records what is different, a commit is the act of actually saving that change to your git repository with the metadata associated. 

## Question 3
#### What is the difference between a branch and a fork?

A branch would exist in the same repository as the original document or code. When working within a branch/creating a branch, the original code would not be affected, but it would have the same history. A fork, however, would be a copy of a document or code in a different repository. This would mean that the original was copied and could be worked on by another user separately. This would also not impact the original document but can be used when there is not full access to the original or it does not live in a repository that you have direct access to but you need to edit/utilize the code. 

## Question 4
#### How are diff and patch related? What’s the difference?

Diff is the content stored in a patch file. While diffs are recorded, indicate the changes made, and can be directly viewed, a patch can be downloaded and used to recreate these changes again.
